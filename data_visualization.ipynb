```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('flood_inter.csv')  # Replace with your CSV path

# ------------------------
# 1. Scatter Plot: Monsoon Intensity vs Flood Probability
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='MonsoonIntensity', y='FloodProbability', alpha=0.6, color='blue')
plt.title('Monsoon Intensity vs Flood Probability')
plt.xlabel('Monsoon Intensity')
plt.ylabel('Flood Probability')
plt.tight_layout()
plt.show()

# ------------------------
# 2. Bar Chart: Top Factors for a Sample Area (first row)
plt.figure(figsize=(12,6))
top_factors = df.iloc[0, :-1]  # All features except FloodProbability
top_factors.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Top Contributing Factors to Floods (Sample Area)')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# ------------------------
# 3. Correlation Heatmap: All Features
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='YlOrRd', cbar=True)
plt.title('Correlation of Flood Risk Factors')
plt.tight_layout()
plt.show()

# ------------------------
# 4. Distribution Plot: Flood Probability
plt.figure(figsize=(8,5))
sns.histplot(df['FloodProbability'], bins=30, kde=True, color='green')
plt.title('Distribution of Flood Probability')
plt.xlabel('Flood Probability')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
