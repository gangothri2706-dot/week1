```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df.drop("FloodProbability", axis=1))

df_scaled = pd.DataFrame(scaled_features, columns=df.columns[:-1])
df_scaled["FloodProbability"] = df["FloodProbability"]

df_scaled.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

X = df_scaled.drop("FloodProbability", axis=1)
y = df_scaled["FloodProbability"]

best_features = SelectKBest(score_func=f_regression, k=5)
fit = best_features.fit(X, y)

# Show feature scores
feature_scores = pd.DataFrame({"Feature": X.columns, "Score": fit.scores_})
print(feature_scores.sort_values(by="Score", ascending=False))
```
